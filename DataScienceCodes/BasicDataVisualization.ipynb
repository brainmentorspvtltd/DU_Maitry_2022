```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
names = ['John','Shawn','Max','Smith','Nick','Steve']
data = {
    "names" : names,
    "dept" : ["IT", "HR", "IT", "HR", "IT", "Admin"],
    "salary" : [56000,25000,70000,55000,45000,30000]
}
```

```python
plt.figure(figsize=(10,6))
plt.bar(data['names'], data['salary'])
plt.title("Salary of Employees")
plt.xlabel("Names")
plt.ylabel("Salary")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.barh(data['names'], data['salary'])
plt.title("Salary of Employees")
plt.ylabel("Names")
plt.xlabel("Salary")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(data['names'], data['salary'], color='red')
plt.title("Salary of Employees")
plt.xlabel("Names")
plt.ylabel("Salary")
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(data['salary'], labels=data['names'], autopct="%1.2f%%", 
        startangle=90, counterclock=False, explode=(0,0.4,0,0,0,0), shadow=True,
       colors=['gold','red','yellow','purple','green','brown'])
plt.title("Salary of Employees")
plt.legend()
plt.show()
```
