```python
import pandas as pd
```

```python
summer_df = pd.read_csv('summer.csv')
winter_df = pd.read_csv('winter.csv')
```

```python
olympics_df = summer_df.append(winter_df)
```

```python
olympics_df.shape
```

```python
olympics_df.head()
```

```python
olympics_df.tail()
```

```python
olympics_df['Athlete'] = olympics_df['Athlete'].str.replace(',','')
```

```python
olympics_df.tail()
```

```python
olympics_df['Athlete'] = olympics_df['Athlete'].str.split().str[::-1].str.join(" ").str.title()
```

```python
olympics_df.head()
```

```python
men_df = olympics_df[olympics_df['Gender'] == 'Men']
women_df = olympics_df[olympics_df['Gender'] == 'Women']
```

```python
pd.value_counts(olympics_df['Gender'])
```

```python
pd.value_counts(men_df['Athlete']).head()
```

```python
pd.value_counts(women_df['Athlete']).head()
```

```python
country_df = olympics_df.groupby(['Country','Medal'])['Sport'].count().reset_index().sort_values(by='Sport', ascending=False)
```

```python
country_df.head()
```

```python
country_df.pivot_table(values='Sport', columns='Medal', index='Country', fill_value=0).sort_values(by='Gold', ascending=False).head()
```
